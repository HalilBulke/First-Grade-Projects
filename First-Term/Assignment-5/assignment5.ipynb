# Part2 (ML Modelling)



We are going to use 5 different ML Regression Models to predict Technical Debt values:

>Linear Regression

>Support Vector Regression

>Decision Tree Regression

>Random Forest Tree Regression

>Multi-layer Perceptron

70% of our data will be used as training set and 30% as test set. Training and test sets will be randomly. The “random_state” variable will have a fixed value so that all models get the same random data. In this way, we will be able to evaluate the success rate of the different models objectively.


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import NearestCentroid

random_state_fixed=53



#This function plots graphs to improve visual confirmation
def plotGraph(y_test,y_pred,regressorName,TD):
  #Plot the Scatter Graph to see results better
    plt.scatter(y_test, y_test,color="r",label="Original values")
    plt.scatter(y_test, y_pred,color="b",label="Predicted values")
    plt.title("Accuracy of %s for %s"%(regressorName,TD))
    plt.legend()
    plt.show()

#Linear Regressor is one of the best choice for the data that is correlated linearly:

def linearRegression(X_train,y_train,X_test,y_test,y1_train,y1_test):
  # Create the regressor: reg_all
    reg = LinearRegression().fit(X_train, y_train)
    reg1 = LinearRegression().fit(X_train, y1_train)
    y_pred = reg.predict(X_test)
    y1_pred = reg1.predict(X_test)
    print("R-Squared Value TD:",r2_score(y_test, y_pred))
    print("Root Mean Squared Error TD:",np.sqrt(mean_squared_error(y_test, y_pred)))
    print("R-Squared Value CDR:",r2_score(y1_test, y1_pred))
    print("Root Mean Squared Error CDR:",np.sqrt(mean_squared_error(y1_test, y1_pred)))
    plotGraph(y_test,y_pred,"Linear regression","TD")
    plotGraph(y1_test,y1_pred,"Linear regression","CDR")
    
    
    
#SVR is also one of the most common regressor that used for regression predictions
def svrRegression(X_train,y_train,X_test,y_test,y1_train,y1_test):
    regressor = SVR(kernel = 'rbf').fit(X_train, y_train.ravel())
    regressor1 = SVR(kernel = 'rbf').fit(X_train, y1_train.ravel())
    y_pred = regressor.predict(X_test)
    y1_pred = regressor1.predict(X_test)
    print("R-Squared Value TD:",r2_score(y_test, y_pred))
    print("Root Mean Squared Error TD:",np.sqrt(mean_squared_error(y_test, y_pred)))
    print("R-Squared Value CDR:",r2_score(y1_test, y1_pred))
    print("Root Mean Squared Error CDR:",np.sqrt(mean_squared_error(y1_test, y1_pred)))
    plotGraph(y_test,y_pred,"SVR Regression","TD-rbf")
    plotGraph(y1_test,y1_pred,"SVR Regression","CDR-rbf")
    regressor = SVR(kernel = 'linear').fit(X_train, y_train.ravel())
    regressor1 = SVR(kernel = 'linear').fit(X_train, y1_train.ravel())
    y_pred = regressor.predict(X_test)
    y1_pred = regressor1.predict(X_test)
    print("R-Squared Value TD:",r2_score(y_test, y_pred))
    print("Root Mean Squared Error TD:",np.sqrt(mean_squared_error(y_test, y_pred)))
    print("R-Squared Value CDR:",r2_score(y1_test, y1_pred))
    print("Root Mean Squared Error CDR:",np.sqrt(mean_squared_error(y1_test, y1_pred)))
    plotGraph(y_test,y_pred,"SVR Regression","TD-linear")
    plotGraph(y1_test,y1_pred,"SVR Regression","CDR-linear")    
        
        
    
def decisionTreeRegression(X_train,y_train,X_test,y_test,y1_train,y1_test):
    regressor = DecisionTreeRegressor(random_state = 53).fit(X_train, y_train.ravel())
    regressor1 = DecisionTreeRegressor(random_state = 53).fit(X_train, y1_train.ravel())
    y_pred = regressor.predict(X_test)
    y1_pred = regressor1.predict(X_test)
    print("R-Squared Value TD:",r2_score(y_test, y_pred))
    print("Root Mean Squared Error TD:",np.sqrt(mean_squared_error(y_test, y_pred)))
    print("R-Squared Value CDR:",r2_score(y1_test, y1_pred))
    print("Root Mean Squared Error CDR:",np.sqrt(mean_squared_error(y1_test, y1_pred)))
    plotGraph(y_test,y_pred,"Decision Tree Regression","TD")
    plotGraph(y1_test,y1_pred,"Decision Tree Regression","CDR")
    



def randomDecTreeRegression(X_train,y_train,X_test,y_test,y1_train,y1_test):
    regressor = RandomForestRegressor(random_state = 53).fit(X_train, y_train.ravel())
    regressor1 = RandomForestRegressor(random_state = 53).fit(X_train, y1_train.ravel())
    y_pred = regressor.predict(X_test)
    y1_pred = regressor1.predict(X_test)
    print("R-Squared Value TD:",r2_score(y_test, y_pred))
    print("Root Mean Squared Error TD:",np.sqrt(mean_squared_error(y_test, y_pred)))
    print("R-Squared Value CDR:",r2_score(y1_test, y1_pred))
    print("Root Mean Squared Error CDR:",np.sqrt(mean_squared_error(y1_test, y1_pred)))
    plotGraph(y_test,y_pred,"Random Decision Tree Regression","TD")
    plotGraph(y1_test,y1_pred,"Random Decision Tree Regression","CDR")


#Multi-layer Perceptron regressor: 'This model optimizes the squared-loss using LBFGS or stochastic gradient descent.'
def mlpRegressor(X_train,y_train,X_test,y_test,y1_train,y1_test):
    regressor = MLPRegressor(random_state=53).fit(X_train, y_train.ravel())
    regressor1 = MLPRegressor(random_state=53).fit(X_train, y1_train.ravel())
    y_pred = regressor.predict(X_test)
    y1_pred = regressor1.predict(X_test)
    print("R-Squared Value TD:",r2_score(y_test, y_pred))
    print("Root Mean Squared Error TD:",np.sqrt(mean_squared_error(y_test, y_pred)))
    print("R-Squared Value CDR:",r2_score(y1_test, y1_pred))
    print("Root Mean Squared Error CDR:",np.sqrt(mean_squared_error(y1_test, y1_pred)))
    plotGraph(y_test,y_pred,"Multi-layer Perceptron Regressor","TD")
    plotGraph(y1_test,y1_pred,"Multi-layer Perceptron Regressor","CDR")
    
```
